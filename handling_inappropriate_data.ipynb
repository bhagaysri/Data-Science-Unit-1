```python
import numpy as np
import pandas as pd
df = pd.read_csv("/content/datasetExample.csv")
df.info()
#in pandas column name.1 represents duplicate column
```

```python
df
```

```python
df.drop_duplicates(inplace = True, ignore_index = True)
df
```

```python
df.drop(['Age_Group.1        '], axis = 1, inplace = True)
df
```

```python
df.columns
```

```python
df = df.rename(columns = { 'Rating(1-5)        ':'Rating(1-5)','Hotel  ': 'Hotel' })
df
```

```python
df.columns
```

```python
df.CustomerID.loc[df.CustomerID < 0] = np.nan
df
```

```python
df.Bill.loc[df.Bill < 0] = np.nan
df
```

```python
df.EstimatedSalary.loc[df.EstimatedSalary < 0] = np.nan
df
```

```python
df['Rating(1-5)'].loc[ (df['Rating(1-5)'] < 1) | (df["Rating(1-5)"] >5)] = np.nan
df
```

```python
df['NoOfPax'].loc[ (df['NoOfPax'] < 1) | (df["NoOfPax"] >20)] = np.nan
df
```

```python
df.Age_Group.unique()
```

```python
df.Hotel.unique()
```

```python
df.Hotel.replace(["Ibys", "ibis", "IbIs"], "Ibis", inplace = True)
df
```

```python
df["FoodPreference"].unique()
df["FoodPreference"].replace(["veg", "Vegetarian"], "Veg", inplace = True)
df["FoodPreference"].replace(["non-veg", "Non-Veg", "non-Veg"], "Non-veg", inplace = True)
df
```

```python
df
```

```python
df['Rating(1-5)'].fillna(method = "ffill", inplace = True)
```

```python
df
```

```python
df['NoOfPax'].fillna(method = "bfill", inplace = True)
df['Bill'].fillna(df['Bill'].mean(), inplace = True)
df['EstimatedSalary'].fillna(df['EstimatedSalary'].mean(), inplace = True)
df
```
