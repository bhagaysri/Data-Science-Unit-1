```python
import numpy as np
import pandas as pd
df = pd.read_csv("/content/employee.csv")
df.info()
df
```

```python
#function to calculate annual income
def cal_sal(Salary):
  inc = 0
  if Salary > 5000:
    inc = Salary*30/100
  else:
    inc = 3000
  return inc
cal_sal(10000)
```

```python
#you can apply cal_sal function on each row for series

df["Increment"] = df.Salary.apply(cal_sal)
df
```
